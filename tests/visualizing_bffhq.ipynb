```python
b = next(iter(dl))
b
```

```python
fig, axes = plt.subplots(1, 5, figsize=(15, 3))
images = b['image']  # shape: (5, C, H, W)
for i in range(5):
    img = images[i].permute(1, 2, 0).cpu().numpy()  # (H, W, C)
    img = ((img +1) / 2).clip(0,1)  # normalize to [0,1]
    axes[i].imshow(img)
    axes[i].set_title(f"Class: {b['class_label'][i].item()}")
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```
